```python
def creatMatrix(n):
    m=[]
    for i in range(n):
        c=[]
        for j in range(n):
            c.append(0)
        m.append(c)
    return m
def printMatrix(m):
    for i in m:
        print(i)
 
 
f=open("Graph04nov.txt","r")
v_e=f.readline().split()
n=int(v_e[0])
m=int(v_e[1])
 
matrix=creatMatrix(n)
 
print()
 
for i in range(m):
    l=f.readline().split()
    u=int(l[0])
    v=int(l[1])
    matrix[u][v]=1
    matrix[v][u]=1 # for undirected
printMatrix(matrix)
 
print()
 
import queue
 
def bfs(G,s,n):
    q=queue.Queue()
    visited=[0]*n
    distance=[0]*n
    visited[s]=1
    distance[s]=0
    q.put(s)
    while not q.empty():
        node=q.get()
        print(node, " ")
        for i in range(n):
            if G[node][i]==1 and visited[i]==0:
                visited[i]=1
                distance[i]=distance[node]+1
                q.put(i)
    print(distance)
 
 
bfs(matrix,0,n)
```

## DFS

```python
def creatMatrix(n):
    m=[]
    for i in range(n):
        c=[]
        for j in range(n):
            c.append(0)
        m.append(c)
    return m
def printMatrix(m):
    for i in m:
        print(i)
 
 
f=open("Graph04nov.txt","r")
v_e=f.readline().split()
n=int(v_e[0])
m=int(v_e[1])
 
matrix=creatMatrix(n)
 
print()
 
for i in range(m):
    l=f.readline().split()
    u=int(l[0])
    v=int(l[1])
    matrix[u][v]=1
    matrix[v][u]=1 # for undirected
printMatrix(matrix)
 
print()
 
import queue
 
def dfs(G,s,n):
    q=queue.LifoQueue()
    visited=[0]*n
    distance=[0]*n
    visited[s]=1
    distance[s]=0
    q.put(s)
    while not q.empty():
        node=q.get()
        print(node, " ")
        for i in range(n):
            if G[node][i]==1 and visited[i]==0:
                visited[i]=1
                #distance[i]=distance[node]+1
                q.put(i)
    print(visited)
 
 
dfs(matrix,0,n)
```
