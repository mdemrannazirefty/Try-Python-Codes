## UCS

```python
import queue

def creatMatrix(n):
    m = []
    for i in range(n):
        c = []
        for j in range(n):
            c.append(0)
        m.append(c)
    return m

def printMatrix(m):
    for i in m:
        print(i)

f = open("wg.txt", "r")
v_e = f.readline().split()
n = int(v_e[0]) 
m = int(v_e[1]) 

matrix = creatMatrix(n)

for i in range(m):
    l = f.readline().split()
    u = int(l[0])
    v = int(l[1])
    weight = int(l[2]) 
    matrix[u][v] = weight

printMatrix(matrix)
print()

def ucs(G, s, n, goal):
    q = queue.PriorityQueue()
    visited = [0] * n
    visited[s] = 1
    q.put((0, s)) 

    while not q.empty():
        gn, node = q.get() 
        print(node, " ") 

        if node == goal:
            return gn  

        for i in range(n):
            if G[node][i] != 9999 and G[node][i]!=0 and visited[i] == 0:  
                visited[i] = 1
                q.put((gn + G[node][i], i)) 

    print(visited)

c=ucs(matrix, 0, n, 4)
print(c)
```
